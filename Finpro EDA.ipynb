## **Matrix Evaluation :** ##



- TP (True Positive): The model correctly predicts policyholders who will renew their policies.
- TN (True Negative): The model correctly predicts policyholders who will not renew their policies.
- FP (False Positive): The model incorrectly predicts that policyholders who will not renew their policies will renew them.
- FN (False Negative): The model incorrectly predicts that policyholders who will renew their policies will not renew them.

There are two types of errors:
- Error type 1 (FP):
Consequences: Waste marketing costs and company resources.
- Error type 2 (FN):
Consequences: Loss of potential candidates who will renew their policies.

However, further analysis is needed by assessing the costs of the errors caused to determine priority treatment for these two types of errors. Using the annual report at PT Asuransi Harta Aman Pratama Tbk for [2022](https://asuransi-harta.co.id/wp-content/uploads/2023/05/AnnualReport2022.pdf), the company got premium revenue in total of 694,123.92 billion rupiah and incurred operating Expenses of 12,607 billion rupiah. Assuming equal value for each insurance product the company offers, we will divide these amounts into seven equal parts. Below is the complete calculation:

1. Calculate the Value of Business Expenses and Auto Insurance Premium Revenue:
- Total Revenue in 2022: 694,123.92 billion rupiah.
- Total operating expenses in 2022: 126.07 billion rupiah.
- Premium Revenue per Insurance Type: 694,123.92 billion / 7 = **99,160.56 billion rupiah**.
- For operating expenses: 126.07 billion/ 7 = **18.009 billion rupiah**.

2. Estimated cost of false positives (FP):
- Based on research by Rodrigo et al. in 2023, false positive rates should not exceed [5%](https://www.frontiersin.org/articles/10.3389/fpsyg.2023.1132128/full), which is a conservative estimate. This rate serves as a basis to assume the proportion of FP.

   So, the estimated FP costs are as follows:

  - **Total costs of FP** = 5% of the value of travel insurance business expenses = 5% * 18.009 billion = **90.043 billion rupiah**.

3. Estimated cost of false negatives (FN):
- Based on research by Rodrigo et al. in 2023, false negative rates should not exceed [20%](https://www.frontiersin.org/articles/10.3389/fpsyg.2023.1132128/full), which is a conservative estimate. This rate serves as a basis to assume the proportion of FN.

   So, the estimated FN costs are as follows:
     - **Total cost of FN**= 20% of auto insurance value = 20% * 99,160.56 billion = **19,832.11 billion rupiah**.


Based on an assessment of the error costs incurred, **the result is that the FN value is greater than the FP**, so as much as possible, We will create a model that can reduce the FN costs so as not to lose potential candidates. According to [(Murphy, 2012)](http://noiselab.ucsd.edu/ECE228/Murphy_Machine_Learning.pdf) recall (sensitivity) becomes a very relevant metric because recall will measure the fraction of actual positive cases identified by the model, which is especially important for larger FNs than FPs. **So, based on considerations, the recall metric will be chosen as the primary metric**.


Additionally, in order to ensure that our model performs effectively in reducing false negatives, **we aim to achieve a recall performance above [80%](https://support.madkudu.com/hc/en-us/articles/4407471867405-Recall-and-Precision-how-is-the-performance-of-a-prediction-model-measured-#:~:text=Here%20is%20an%20approximative%20rule,%3A%20Recall%20%3D%2060%2D70%25)**. Achieving this level of recall ensures that our model is sufficiently sensitive in identifying policyholders who are likely to renew their policies, thereby minimizing the risk of overlooking these potential candidates.

##### **Distribution of Numerical Data With The Anderson-Darling Test**

```python
# Selecting numeric columns from 'dfai'
numeric_columns_dfai = dfai.select_dtypes(include=['number']).columns

# Conducting the Anderson-Darling Test for each numeric column in 'dfai'
test_results_dfai = []
for column in numeric_columns_dfai:
    stat, critical_values, _ = stats.anderson(dfai[column], dist='norm')
    conclusion = 'Not normally distributed' if stat > critical_values[2] else 'Normally distributed'
    test_results_dfai.append([column, stat, critical_values[2], conclusion])

# Creating a table of the results
headers_dfai = ['Column', 'Statistic', 'Critical Value (5%)', 'Conclusion']
formatted_table_dfai = tabulate(test_results_dfai, headers=headers_dfai, tablefmt='pretty')
print(formatted_table_dfai)
```

##### **Point-Biserial Correlation**

After establishing a foundational understanding of our data distributions, we progressed to exploring the relationships between the numeric features and the binary 'Response' target variable using point-biserial correlation. The point-biserial correlation assesses relationships between continuous variables (such as 'Customer Lifetime Value,' 'Income,' 'Monthly Premium Auto,' 'Months Since Last Claim,' 'Months Since Policy Inception,' 'Number of Open Complaints,' 'Number of Policies', and 'Total Claim Amount') and the binary 'Response' variable, which indicates whether a customer decided to renew their auto insurance policy. This method is chosen for its effectiveness in quantifying the association between continuous and dichotomous variables, offering a clear range (-1 to 1) for interpreting the correlation's strength and direction. Its application is ideal for datasets with naturally dichotomous outcomes, such as our dataset where 'Response' represents a clear binary choice (renewal or non-renewal), making it a relevant and insightful tool in our exploratory data analysis ([Laerd Statistics](https://statistics.laerd.com/spss-tutorials/point-biserial-correlation-using-spss-statistics.php), [Statistics Solutions](https://www.statisticssolutions.com/point-biserial-correlation-coefficient/)).

```python
# Make a copy of the dataframe 'dfai'
dfx = dfai.copy()

# Identify numeric features in the dataset
numeric_features = dfx.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Compute point-biserial correlation for each numeric feature with the target 'Response_num'
point_biserial_correlations = {}
for feature in numeric_features:
    if feature != 'Response':
        # Ensure that both series do not contain NaN values and have at least 2 data points before calculation
        valid_data = dfx[[feature, 'Response']].dropna()
        if len(valid_data) >= 2:
            correlation, _ = pointbiserialr(valid_data[feature], valid_data['Response'])
            point_biserial_correlations[feature] = correlation

# Converting the results to a DataFrame
correlation_table = pd.DataFrame.from_dict(point_biserial_correlations, orient='index', columns=['Correlation'])

# Sort the table by correlation values
correlation_table.sort_values(by='Correlation', ascending=False, inplace=True)

# Converting the correlation table to a format suitable for tabulate
correlation_table_tabulate = correlation_table.reset_index()
correlation_table_tabulate.columns = ['Feature', 'Correlation']

# Displaying the table using tabulate
print(tabulate(correlation_table_tabulate, headers='keys', tablefmt='grid', showindex=False))
```

#### **Data Correlation**

##### **Chi Square Test**

The [Chi-square test](https://pubmed.ncbi.nlm.nih.gov/28295394/#:~:text=URL%3A%20https%3A%2F%2Fpubmed) is a robust statistical method for determining if there is a significant association between two categorical variables. It is particularly relevant for our auto insurance dataset because it can test whether the distribution of policy renewals across different categories, such as employment status, renewal offer types, or vehicle classes, is due to chance or signifies a natural correlation. This test is appropriate for our analysis as it does not require the data to follow a normal distribution. It is suitable when the variables are nominal, as with most categorical data in our dataset. By applying the Chi-square test, we can identify which factors might influence a customer's decision to renew their auto insurance policy, providing valuable insights for strategic decision-making and policy design.

```python
# Identify categorical features
categorical_features = dfai.select_dtypes(include=['object']).columns.tolist()
categorical_features.remove('Customer')  # Remove 'Customer' from the list

# Initialize a dictionary to store Chi-square test results
chi_square_results = {}

# Conducting Chi-square tests
for feature in categorical_features:
    if feature != 'Response':  # Exclude the target variable
        contingency_table = pd.crosstab(dfai[feature], dfai['Response'])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        chi_square_results[feature] = {'Chi2 Statistic': chi2, 'p-value': p}

# Convert the results to a DataFrame for better visualization
chi_square_results_df = pd.DataFrame(chi_square_results).T

# Sorting the DataFrame by p-value
chi_square_results_df.sort_values(by='p-value', inplace=True)

# Format p-values to scientific notation with 3 decimal places
chi_square_results_df['p-value'] = chi_square_results_df['p-value'].apply(lambda x: '{:.3e}'.format(x))

# Apply coloring to p-values less than 0.05
styled_chi_square_results_df = chi_square_results_df.style.applymap(
    lambda x: 'color: red' if float(x) < 0.05 else 'color: black', subset=pd.IndexSlice[:, ['p-value']]
)

# Display the results with styles
styled_chi_square_results_df
```

#### **Bivariate Correlation**

##### **Cross Tabulation**

[Cross tabulation](https://www.alchemer.com/resources/blog/cross-tabulation/#) is a valuable technique for bivariate correlation analysis as it helps to clarify the relationship between two categorical variables, making complex data sets more manageable and insights more discernible. This approach is particularly useful in market research or survey analysis, where it can reveal patterns and connections that inform strategic decisions.

```python
# Identify categorical features in the dataset
categorical_features = dfai.select_dtypes(include=['object']).columns.tolist()
categorical_features.remove('Customer')  # Remove 'Customer' from the list

# Perform cross-tabulation for each categorical feature against 'Response'
cross_tab_results = {}
for feature in categorical_features:
    if feature != 'Response':  # Exclude the target variable
        cross_tab = pd.crosstab(dfai[feature], dfai['Response'])
        cross_tab_results[feature] = cross_tab

# Display the cross-tabulation results
for feature, cross_tab in cross_tab_results.items():
    print(f"Cross-tabulation for {feature}:\n{cross_tab}\n")
```
