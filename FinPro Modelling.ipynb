## **Matrix Evaluation :** ##



- TP (True Positive): The model correctly predicts policyholders who will renew their policies.
- TN (True Negative): The model correctly predicts policyholders who will not renew their policies.
- FP (False Positive): The model incorrectly predicts that policyholders who will not renew their policies will renew them.
- FN (False Negative): The model incorrectly predicts that policyholders who will renew their policies will not renew them.

There are two types of errors:
- Error type 1 (FP):
Consequences: Waste marketing costs and company resources.
- Error type 2 (FN):
Consequences: Loss of potential candidates who will renew their policies.

However, further analysis is needed by assessing the costs of the errors caused to determine priority treatment for these two types of errors. Using the annual report at PT Asuransi Harta Aman Pratama Tbk for [2022](https://asuransi-harta.co.id/wp-content/uploads/2023/05/AnnualReport2022.pdf), the company got premium revenue in total of 694,123.92 billion rupiah and incurred operating Expenses of 12,607 billion rupiah. Assuming equal value for each insurance product the company offers, we will divide these amounts into seven equal parts. Below is the complete calculation:

1. Calculate the Value of Business Expenses and Auto Insurance Premium Revenue:
- Total Revenue in 2022: 694,123.92 billion rupiah.
- Total operating expenses in 2022: 126.07 billion rupiah.
- Premium Revenue per Insurance Type: 694,123.92 billion / 7 = **99,160.56 billion rupiah**.
- For operating expenses: 126.07 billion/ 7 = **18.009 billion rupiah**.

2. Estimated cost of false positives (FP):
- Based on research by Rodrigo et al. in 2023, false positive rates should not exceed [5%](https://www.frontiersin.org/articles/10.3389/fpsyg.2023.1132128/full), which is a conservative estimate. This rate serves as a basis to assume the proportion of FP.

   So, the estimated FP costs are as follows:

  - **Total costs of FP** = 5% of the value of travel insurance business expenses = 5% * 18.009 billion = **90.043 billion rupiah**.

3. Estimated cost of false negatives (FN):
- Based on research by Rodrigo et al. in 2023, false negative rates should not exceed [20%](https://www.frontiersin.org/articles/10.3389/fpsyg.2023.1132128/full), which is a conservative estimate. This rate serves as a basis to assume the proportion of FN.

   So, the estimated FN costs are as follows:
     - **Total cost of FN**= 20% of auto insurance value = 20% * 99,160.56 billion = **19,832.11 billion rupiah**.


Based on an assessment of the error costs incurred, **the result is that the FN value is greater than the FP**, so as much as possible, We will create a model that can reduce the FN costs so as not to lose potential candidates. According to [(Murphy, 2012)](http://noiselab.ucsd.edu/ECE228/Murphy_Machine_Learning.pdf) recall (sensitivity) becomes a very relevant metric because recall will measure the fraction of actual positive cases identified by the model, which is especially important for larger FNs than FPs. **So, based on considerations, the recall metric will be chosen as the primary metric**.


Additionally, in order to ensure that our model performs effectively in reducing false negatives, **we aim to achieve a recall performance above [80%](https://support.madkudu.com/hc/en-us/articles/4407471867405-Recall-and-Precision-how-is-the-performance-of-a-prediction-model-measured-#:~:text=Here%20is%20an%20approximative%20rule,%3A%20Recall%20%3D%2060%2D70%25)**. Achieving this level of recall ensures that our model is sufficiently sensitive in identifying policyholders who are likely to renew their policies, thereby minimizing the risk of overlooking these potential candidates.

### **Handling Outlier in Customer Lifetime Value**

```python
# Calculate the Interquartile Range (IQR) for Customer Lifetime Value
Q1 = dfai['Customer Lifetime Value'].quantile(0.25)
Q3 = dfai['Customer Lifetime Value'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds using IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers before Winsorizing
outliers_before = dfai[(dfai['Customer Lifetime Value'] < lower_bound) | 
                       (dfai['Customer Lifetime Value'] > upper_bound)]

# Calculating the percentage of outliers before Winsorizing
outlier_count_before = outliers_before.shape[0]
total_count = dfai.shape[0]
outlier_percentage_before = (outlier_count_before / total_count) * 100

# Checking the distribution of 'Response' among the outliers before Winsorizing
response_distribution_outliers_before = outliers_before['Response'].value_counts()

# Apply Winsorizing to 'Customer Lifetime Value' based on IQR boundaries
dfai['Customer Lifetime Value Winsorized'] = winsorize(dfai['Customer Lifetime Value'], 
                                                       limits=(None, None), 
                                                       inplace=False)
# Update the winsorized column with IQR boundaries
dfai['Customer Lifetime Value Winsorized'] = dfai['Customer Lifetime Value Winsorized'].clip(lower=lower_bound, upper=upper_bound)

# Identifying outliers after Winsorizing (should be none or very few)
outliers_after = dfai[(dfai['Customer Lifetime Value Winsorized'] < lower_bound) | 
                      (dfai['Customer Lifetime Value Winsorized'] > upper_bound)]

# Calculating the percentage of outliers after Winsorizing
outlier_count_after = outliers_after.shape[0]
outlier_percentage_after = (outlier_count_after / total_count) * 100

# Checking the distribution of 'Response' among the outliers after Winsorizing
response_distribution_outliers_after = outliers_after['Response'].value_counts() if not outliers_after.empty else pd.Series([])

# Descriptive statistics for the original and Winsorized 'Customer Lifetime Value'
original_stats = dfai['Customer Lifetime Value'].describe()
winsorized_stats = dfai['Customer Lifetime Value Winsorized'].describe()

# Drop the original 'Customer Lifetime Value' column
dfai = dfai.drop('Customer Lifetime Value', axis = 1)

# Output the descriptive statistics and outlier information
print("Original 'Customer Lifetime Value' statistics:\n", original_stats)
print("\nWinsorized 'Customer Lifetime Value' statistics:\n", winsorized_stats)
print("\nOutlier percentage before Winsorizing:", outlier_percentage_before)
print("Outlier count before Winsorizing:", outlier_count_before)
print("Response distribution among outliers before Winsorizing:\n", response_distribution_outliers_before)
print("\nOutlier percentage after Winsorizing:", outlier_percentage_after)
print("Outlier count after Winsorizing:", outlier_count_after)
print("Response distribution among outliers after Winsorizing:\n", response_distribution_outliers_after)
```

## **Handling Rare Label**

Rare labels in a dataset refer to categories within a categorical variable that occur infrequently. These labels represent a small percentage of the data and are distinct from the more commonly occurring categories.

```python
from tabulate import tabulate

# Selecting categorical columns without customer
categorical_columns = dfai.select_dtypes(include=['object']).columns.tolist()

# Remove 'Customer' from the list
categorical_columns.remove('Customer')

# Calculate the frequency distribution for each categorical variable in percentage
label_frequency_percent = {col: dfai[col].value_counts(normalize=True) * 100 for col in categorical_columns}

# Now, we will create a nested list of the data to be able to use it with tabulate directly.
# The outer list corresponds to the rows of the table and the inner lists correspond to the columns of each row.
data_for_tabulate = []

# Iterate over the dictionary to populate the data for tabulate
for variable, freq_series in label_frequency_percent.items():
    data_for_tabulate.append([variable, freq_series.index[0], f"{freq_series.iloc[0]:.2f}%"])
    for idx in range(1, len(freq_series)):
        data_for_tabulate.append(["", freq_series.index[idx], f"{freq_series.iloc[idx]:.2f}%"])

# Display the frequency distribution in percentage using tabulate
print(tabulate(data_for_tabulate, headers=["Variable", "Category", "Frequency (%)"], tablefmt="pipe"))
```

### SMOTE Setup

[The Synthetic Minority Over-sampling Technique (SMOTE)](https://arxiv.org/abs/1106.1813) is a widely recognized method for addressing class imbalance in datasets. SMOTE creates synthetic samples from the minority class to promote equal representation. It achieves this by interpolating between existing minority class instances to generate new, synthetic instances. By employing SMOTE during benchmarking, we can assess the impact of a balanced class distribution on the performance of models like Naive Bayes, Logistic Regression, and Ridge Classifier. This insight is pivotal for understanding how each model handles a more balanced dataset and maintains performance metrics, especially for minority class predictions.

```python
smote_experiment = ClassificationExperiment()
smote_experiment.setup(
    experiment_name='SMOTE Resampling',
    custom_pipeline=[
        ('preprocessing', TransformerWrapper(transformer=transformers)),
        ('resampling', TransformerWrapper(transformer=FixImbalancer(SMOTE(random_state=random_seed)))),
    ],
    **configs
)
```

### NearMiss Setup

[Near-Miss](https://arxiv.org/abs/1106.1813) is an algorithm that can help in balancing an imbalanced dataset. It can be grouped under undersampling algorithms and is an efficient way to balance the data. The algorithm does this by looking at the class distribution and randomly eliminating samples from the larger class. When two points belonging to different classes are very close to each other in the distribution, this algorithm eliminates the datapoint of the larger class thereby trying to balance the distribution. 

```python
nearmiss_experiment = ClassificationExperiment()
nearmiss_experiment.setup(
    experiment_name='NearMiss Resampling',
    custom_pipeline=[
        ('preprocessing', TransformerWrapper(transformer=transformers)),
        ('resampling', TransformerWrapper(transformer=FixImbalancer(NearMiss(version=3)))),
    ],
    **configs
)
```

These methods are aptly named because they “wrap” around the machine learning algorithm in question and iteratively evaluate different combinations of features to determine which subset results in the best model performance.

### Model Limitation

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

predicted_probabilities = final_model.predict_proba(get_config('X_test_transformed'))[:, 1]

y_test = get_config('y_test_transformed') 

bins = np.linspace(0, 1, 11)
bin_indices = np.digitize(predicted_probabilities, bins) - 1

results_data = []

for i in range(len(bins) - 1):
    in_bin = bin_indices == i
    actual_in_bin = y_test[in_bin]
    predictions_in_bin = (predicted_probabilities[in_bin] > 0.5).astype(int)

    accuracy = accuracy_score(actual_in_bin, predictions_in_bin)
    precision = precision_score(actual_in_bin, predictions_in_bin, zero_division=0)
    recall = recall_score(actual_in_bin, predictions_in_bin, zero_division=0)
    f1 = f1_score(actual_in_bin, predictions_in_bin, zero_division=0)
    
    range_label = f"{bins[i]*100:.0f}% - {bins[i+1]*100:.0f}%"
    results_data.append({
        'Range Probability': range_label,
        'Accuracy' : accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    })

results_df = pd.DataFrame(results_data)
results_df
```

## Model Interpretation

XGBoost is a robust machine learning algorithm for regression and classification problems, an extension of gradient-boosted decision trees. It is optimized for performance and is capable of handling large datasets efficiently. XGBoost stands for eXtreme Gradient Boosting and is particularly well-suited for structured data.

The core of XGBoost is an ensemble of decision trees, where each tree is built to correct the errors made by the previous ones. It uses a gradient descent algorithm to minimize the loss when adding new models. This algorithm involves sequentially creating and adding trees to the model until no further improvements can be made. XGBoost also incorporates regularization, which helps to prevent overfitting by penalizing more complex models.

The model calculates the likelihood of a particular outcome, effectively distinguishing between the possible classes. This makes it a robust choice for various tasks, from classification to ranking.

However, XGBoost has its limitations. While it handles a mix of numerical and categorical features well, it may still be challenged by extremely noisy or unstructured data. Additionally, it can overfit without careful tuning and regularization, especially when the data is not diverse enough or too small. XGBoost's complexity can also lead to longer training times compared to simpler models, and understanding the contributions of individual features can be less intuitive, often requiring additional tools like SHAP values for interpretation. Despite these challenges, when used appropriately, XGBoost is a highly effective and versatile tool for predictive modeling.
